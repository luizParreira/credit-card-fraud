## Sampling

To sample the data, there a few methods we can use, such as over sampling, under sampling and SMOTE. We are going to be testing our results on all of those methods, to see for ourselves if they can actually improve our results. 

1. To start off, we will do a simple random under-sampling
2. Secondly, use a simple methods such as random under/over sampling.
2. Then, we intend to use more advanced one, called SMOTE

```python
 # Breaking the dataset into X and Y
    
    
y = data['Class']
X = data.drop('Class', axis = 1)

def random_undersampling(df):
    shuffled_data = df.sample(frac=1)
    fraud_data = shuffled_data.loc[shuffled_data['Class'] == 1]
    non_fraud_data = shuffled_data.loc[shuffled_data['Class'] == 0][:fraud_data.shape[0]]

    normal_distributed_df = pd.concat([fraud_data, non_fraud_data])

    return normal_distributed_df.sample(frac=1, random_state=random_state)

new_data = random_undersampling(data)
new_y = new_data['Class']
new_X = new_data.drop('Class', axis = 1)
```

## Evaluating model performance


This is one of the most important parts of the whole process. This is where we align business objectives' with the output that we want form our model.

In our case, we have a fraud detection problem. Fraud, is an expensive problem for any company that has to deal with financial transactions. It hurts their finances and their capacity to invest and innovate within their business. With that in mind, its more important to identify the fraudulen transactions than the non fraudulent transactions, meaning our metric has to incorporate that fact. One metric we can use to evaluate such models are ones that take into consideration both **precision** and **recall**, such as **F-beta score**:

$$ F_{\beta} = (1 + \beta^2) \cdot \frac{precision \cdot recall}{\left( \beta^2 \cdot precision \right) + recall} $$


In particular, when $\beta = 0.5$, more emphasis is placed on precision. This is called the **F$_{0.5}$ score** (or F-score for simplicity).

Lets take a quick look at **precision** and **recall**:

**Precision** tells us what proportion of transactions we classified as fraud, actually were fraud.
It is a ratio of true positives(transactions classified as fraud, and which are actually frauds) to all positives(all transactions classified as fraud, irrespective of whether that was the correct classification), in other words it is the ratio of

`[True Positives/(True Positives + False Positives)]`


**Recall(sensitivity)** tells us what proportion of transactions that actually were fraud were classified by us as fraud.
It is a ratio of true positives(transactions classified as fraud, and which are actually fraud) to all the transactions that were actually fraud, in other words it is the ratio of

`[True Positives/(True Positives + False Negatives)]`

 # Initial model evaluation
 
 We have selected a few models to test out on this dataset, in order to be able to establish a baseline that we can then try and improve by doing some feature engineering, to identify which features we can drop. Also, its important to choose diverse models that base their classifications on distinct ideas, in order to be able to better analyze the data available to us.
 
 

```python
# TODO: Import two metrics from sklearn - fbeta_score and accuracy_score
from sklearn.metrics import fbeta_score, accuracy_score
from sklearn.model_selection import cross_val_score
from time import time


def train_predict(model, user_data): 
    '''
    inputs:
       - learner: the learning algorithm to be trained and predicted on
       - data: (X_train, X_test, y_train, y_test): Data structure containing training and testing datasets
    '''
    
    (X_trainn, X_testt, y_trainn, y_testt) = user_data
    results = {}
    
    start = time()
    learner = model.fit(X_trainn, y_trainn)
    end = time() 
    
    training_score = cross_val_score(model, X_trainn, y_trainn, cv=5)
    print()
    print("Classifiers: ", model.__class__.__name__, "Has a training score of", round(training_score.mean(), 2) * 100, "% accuracy score")
    print()
    results['train_time'] = end - start
        

    start = time()
    predictions_test = model.predict(X_testt)
    predictions_train = model.predict(X_trainn)
    end = time()
    
    results['pred_time'] = end - start
    results['acc_train'] = accuracy_score(y_trainn, predictions_train)
    results['acc_test'] = accuracy_score(y_testt, predictions_test)
    results['f_train'] = fbeta_score(y_trainn, predictions_train, average='weighted', beta=0.5)        
    results['f_test'] = fbeta_score(y_testt, predictions_test, average='weighted', beta=0.5)
       
    print("{} trained on {} samples.".format(model.__class__.__name__, X_trainn.shape[0]))
        
    return results
```

```python

# TODO: Import the three supervised learning models from sklearn
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier


sampled_data = [
    _train_test_split(new_X, new_y)
]

classifiers = [
    xgb.XGBClassifier(random_state = random_state),
    GaussianNB(),
    svm.SVC(random_state = random_state, gamma='auto'),
    LogisticRegression(random_state=random_state, solver='lbfgs'),
    KNeighborsClassifier()
    
]
def run_clfs(classifiers, sampled_data):
    # Collect results on the learners
    results = {}
    for clf in classifiers:
        clf_name = clf.__class__.__name__
        results[clf_name] = {}
        for i, samples in enumerate(sampled_data):
            results[clf_name][i] = train_predict(clf, samples)
    return results
results = run_clfs(classifiers, sampled_data)
```

```python
def report_results(resultss):
    for clf_name, result in resultss.items():
        print(clf_name)
        print()
        for i, values in result.items():
            print('Trained in {} seconds'.format(values['train_time']))
            print('Tested in {} seconds'.format(values['pred_time']))
            print('Got a {} % accuracy on training'.format(values['acc_train'] * 100))
            print('Got a {} % accuracy on testing'.format(values['acc_test'] * 100))
            print('Got a {} % f-score on training'.format(values['f_train'] * 100))
            print('Got a {} % f-score on testing'.format(values['f_test'] * 100))
            print()
            
report_results(results)
```

## Outlier detection

Earlier we looked at the distribution of amounts and time, and we noticed that the data was largely distributed in small transactions, while theere were a few large value transactions that brought the distribution all the way up to 25000. In order for our models to better generalize, we will use Tukey's Method for identfying outliers: An *outlier step* is calculated as 1.5 times the interquartile range (IQR). A data point with a feature that is beyond an outlier step outside of the IQR for that feature is considered abnormal. Once we remove all the outliers from the data, we are going to run the undersampling algorithm once more, on the newly transformed data.

```python
all_outliers = []
for feature in X.keys():
  
    Q1 = np.percentile(X[feature], 25)
    Q3 = np.percentile(X[feature], 75)
    
    step = 1.5 * (Q3 - Q1)
    
    outliers = X[~((X[feature] >= Q1 - step) & (X[feature] <= Q3 + step))]
    all_outliers.append(outliers.index.values)
    
all_outliers = [item for sublist in all_outliers for item in sublist]
good_data = data.drop(data.index[all_outliers]).reset_index(drop = True)
```

```python

new_data = random_undersampling(good_data)


def get_x_and_y(df):
    YY = df['Class']
    XX = df.drop('Class', axis = 1)
    
    return XX, YY

new_X, new_y = get_x_and_y(new_data)
X, y = get_x_and_y(good_data)

chosen_classifiers = [
    xgb.XGBClassifier(random_state = random_state),
    LogisticRegression(random_state=random_state, solver='lbfgs'),
]

sampled_data = [
    _train_test_split(X, y)
]

results = run_clfs(chosen_classifiers, sampled_data)

```

### PCA

In order to optimize our models even further, we are going to use a technique called Principal Component Analysis (PCA) to draw conclusions of the underlying structure of the data presented to us. PCA is used to calculate the dimensions which best maximize variance within the dataset, it shows us which combinations of features best describe our transactions.

```python

import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.decomposition import PCA

def pca_results_plot(df, pca):
    '''
    Create a DataFrame of the PCA results
    Includes dimension feature weights and explained variance
    Visualizes the PCA results
    '''

    # Dimension indexing
    dimensions = dimensions = ['Dimension {}'.format(i) for i in range(1,len(pca.components_)+1)]

    # PCA components
    components = pd.DataFrame(np.round(pca.components_, 4), columns = df.keys())
    components.index = dimensions

    # PCA explained variance
    ratios = pca.explained_variance_ratio_.reshape(len(pca.components_), 1)
    variance_ratios = pd.DataFrame(np.round(ratios, 4), columns = ['Explained Variance'])
    variance_ratios.index = dimensions

    # Create a bar plot visualization
    fig, ax = plt.subplots(figsize = (14,8))

    # Plot the feature weights as a function of the components
    components.plot(ax = ax, kind = 'bar');
    ax.set_ylabel("Feature Weights")
    ax.set_xticklabels(dimensions, rotation=0)


    # Display the explained variance ratios
    for i, ev in enumerate(pca.explained_variance_ratio_):
        ax.text(i-0.40, ax.get_ylim()[1] + 0.05, "Explained Variance\n          %.4f"%(ev))

    # Return a concatenated DataFrame
    return pd.concat([variance_ratios, components], axis = 1)


def apply_pca(n_components, the_data):

    n_components  = 10
    # TODO: Apply PCA by fitting the good data with the same number of dimensions as features
    pca = PCA(n_components=n_components).fit(the_data)


    reduced_data = pca.transform(the_data)

    # Generate PCA results plot
    pca_results = pca_results_plot(the_data, pca)
    dimentions = []
    for i in range(1, n_components + 1):
        dimentions.append('Dimension ' + str(i))
        
    reduced_data = pd.DataFrame(reduced_data, columns = dimentions)
    return reduced_data, pca_results

reduced_data, pca_results = apply_pca(10, X)
```
